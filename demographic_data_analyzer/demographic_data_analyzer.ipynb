What is the average age of men?


```python
round(df[df["sex"]=="Male"]["age"].mean(),1)
```

What is the percentage of people who have a Bachelor's degree?

```python
round((df["education"]=="Bachelors").mean()*100,1)
```

What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

```python
round((df[df["education"].isin(["Bacholers","Masters","Doctorate"])]["salary"]==">50K").mean()*100,1)
```

What percentage of people without advanced education make more than 50K?

```python
round((df[~(df["education"].isin(["Bacholers","Masters","Doctorate"]))]["salary"]==">50K").mean()*100,1)
```

What is the minimum number of hours a person works per week?

```python
df["hours-per-week"].min()
```

What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
round((df[df['hours-per-week']==df["hours-per-week"].min()]["salary"]==">50K").mean()*100,1)
```

What country has the highest percentage of people that earn >50K and what is that percentage?

Percentage of rich=(Total people in that countryRich people in that country​)×100


```python
country_rich=(
    df[df["salary"] == ">50K"]["native-country"].value_counts()
    /
    df["native-country"].value_counts()
) * 100
```

```python
highest_earning_country = country_rich.idxmax()
round(country_rich.max(), 1)
```
